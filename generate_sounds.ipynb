```python
import scaper, os
import numpy as np
```

```python
# Scaper Settings
n_soundscapes = 2000
n_start = 1000
n_stop = 2000
duration = 10
outfolder = 'soundscape/'
jamsfolder = 'jams/'
txtfolder = 'txt/'



fg_folder = os.path.expanduser("foreground")
bg_folder = os.path.expanduser("background")
```

```python
%%time
# Generate n audio files without any events
for i in range(n_start, n_stop):
    print('Generating normal soundclip: {:d}/{:d}'.format(i+1, n_soundscapes))
    
    # create a scaper
    sc = scaper.Scaper(duration, fg_folder, bg_folder)
    sc.protected_labels = []
    sc.ref_db = 0

    # Add background
    sc.add_background(label = ("const", "rainforest_day"),
                     source_file = ("choose", []),
                     source_time = ("uniform", 0, 30 - duration))
    
    audiofile = outfolder + f'normal_{i}.wav'
    jamsfile = jamsfolder + f'normal_{i}.jams'
    txtfile = txtfolder + f'normal_{i}.txt'
    sc.generate(audiofile, jamsfile,
                allow_repeated_label = True,
                allow_repeated_source = True,
                disable_sox_warnings = True,
                no_audio = False,
                txt_sep = ',',
                txt_path = txtfile)
```

```python
%%time
# Generate n soundscapes
for i in range(n_start, n_stop):
    print('Generating soundscape: {:d}/{:d}'.format(i+1, n_soundscapes))
    
    # create a scaper
    sc = scaper.Scaper(duration, fg_folder, bg_folder)
    sc.protected_labels = []
    sc.ref_db = 0

    # Add background
    sc.add_background(label = ("const", "rainforest_day"),
                     source_file = ("choose", []),
                     source_time = ("uniform", 0, 30 - duration))
    
    # add random number of foreground events
    n_events = np.random.randint(1, 5)
    
    for _ in range(n_events):
        event_time = np.random.randint(0, duration)
        event_duration = duration - event_time
        sc.add_event(label = ("const", "chopping"),
                     source_file = ("choose", []),
                     source_time = ("uniform", 0, 30 - duration),
                     event_time = ("uniform", 0, event_time),
                     event_duration = ("uniform", 1, event_duration),
                     snr = ("uniform", -5, 0),
                     pitch_shift = ("uniform", -15, 15),
                     time_stretch = None)
    
    audiofile = outfolder + f'soundscape_{i}.wav'
    jamsfile = jamsfolder + f'soundscape_{i}.jams'
    txtfile = txtfolder + f'soundscape_{i}.txt'
    
    sc.generate(audiofile, jamsfile,
                allow_repeated_label = True,
                allow_repeated_source = True,
                disable_sox_warnings = True,
                no_audio = False,
                txt_sep = ',',
                txt_path = txtfile)
```
