```python
import librosa.display, os, gc
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generates the Images / Spectrograms from the audio files - One N number of normal sounds, another N number of forest + chainsaw sounds.
def extract_spectrogram(fname, iname):
    audio, sr = librosa.load(fname, res_type = 'kaiser_best')
    S = librosa.feature.melspectrogram(audio, sr = sr, n_mels = 128)
    log_S = librosa.power_to_db(S, ref = np.max)
    fig = plt.figure(figsize = [1, 1])
    ax = fig.add_subplot(111)
    fig.subplots_adjust(left = 0,right = 1,bottom = 0,top = 1)
    ax.axis("off")
    ax.axis("tight")
    plt.margins(0)
    librosa.display.specshow(log_S, sr = sr)
    fig.savefig(iname, dpi = 200, pad_inches = 0) # You can set a high resolution instead of 100 dpi. e.g. 100 dpi = 100 x 100.
    plt.close(fig)
    plt.close('all')
    del audio, S, log_S, ax, fig
```

```python
%%time
samples_folder = "soundscape/"
images_folder = "images/"
already = os.listdir(images_folder)
d = os.listdir(samples_folder)
d.sort()
for i, f in enumerate(d[1650:]):
    print('Generating spectrogram {:d} for: {:s}'.format(i+1, f))
    extract_spectrogram(samples_folder+f, f"{images_folder}/{f.replace('.wav', '.png')}")
```
